# INSIGHT

# 1. Hari Paling Sibuk — Bar Chart
Tujuan: Mengetahui hari dengan jumlah order terbanyak sehingga perusahaan bisa menentukan waktu terbaik untuk promosi, stok barang, dan penjadwalan operasional.

```python
import matplotlib.pyplot as plt

# Menghitung jumlah order per hari
busy_day = df['order_dow'].value_counts().sort_index()

# Membuat visualisasi
plt.figure(figsize=(8,5))

sns.barplot(
    x=busy_day.index,
    y=busy_day.values
)

# Judul dan label
plt.title("Hari Paling Sibuk")
plt.xlabel("Hari")
plt.ylabel("Jumlah Order")

# Optional: ganti angka jadi nama hari
plt.xticks(
    ticks=range(7),
    labels=['Minggu','Senin','Selasa','Rabu','Kamis','Jumat','Sabtu']
)

plt.show()
```

# Insight
*   Hari 0 (Minggu) memiliki transaksi tertinggi.
*   Banyak pelanggan berbelanja saat akhir pekan.

Rekomendasi
*   Tambah stok menjelang akhir pekan.
*   Jalankan promo weekend.



# 2. Rata-rata Customer Melakukan Re-order — Histogram
Tujuan: Melihat distribusi rata-rata reorder customer untuk memahami loyalitas pelanggan dan kebiasaan repeat order.

```python
import matplotlib.pyplot as plt

customer_reorder = df.groupby('user_id')['reordered'].mean()

plt.figure(figsize=(8,5))
plt.hist(customer_reorder, bins=20)

mean_reorder = customer_reorder.mean()

plt.axvline(
    mean_reorder,
    linestyle='--',
    linewidth=2,
    label=f'Rata-rata = {mean_reorder:.2f}'
)

plt.title('Rata-rata Customer Melakukan Re-order')
plt.xlabel('Reorder Rate')
plt.ylabel('Jumlah Customer')
plt.legend()

plt.show()
```

# Insight
 *   Jika mayoritas customer berada di rentang 0.6–0.8, berarti banyak pelanggan yang sering membeli ulang produk yang sama.
*   Jika mayoritas berada di rentang 0–0.3, berarti pelanggan cenderung mencoba produk baru dan loyalitas masih rendah.





# 3. Produk Organik vs Non Organik — Pie Chart
Tujuan: Mengetahui minat customer terhadap produk organik dibanding non-organik untuk membantu strategi produk dan pemasaran.

```python
import matplotlib.pyplot as plt

#memisahkan produk organik dan non organik
organic = df[
    df['product_name']
    .str.contains('Organic', case=False, na=False)
]

non_organic = df[
    ~df['product_name']
    .str.contains('Organic', case=False, na=False)
]

# Jumlah perbandingan organik dan non organik pie chart
labels = ['Organic', 'Non-Organic']
sizes = [len(organic), len(non_organic)]

# Plot
plt.figure(figsize=(7,7))

plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%'
)

plt.title("Organic vs Non-Organic Products")

plt.show()
```

# Insight
*   Jika Non Organic mendominasi, berarti mayoritas pelanggan masih memilih produk reguler.
*   Jika Organic memiliki persentase tinggi, menunjukkan adanya minat besar terhadap produk sehat dan premium.



# 4. Customer Activity Timeline — Line Chart
Tujuan: Melihat pola aktivitas customer berdasarkan jam belanja untuk menentukan prime time transaksi.

```python
import matplotlib.pyplot as plt

# Jumlah order unik per jam
timeline = (
    df.groupby('order_hour_of_day')['order_id']
      .nunique()
      .sort_index()
)

plt.figure(figsize=(14, 5))

# Line chart
plt.plot(
    timeline.index,
    timeline.values,
    marker='o',
    linestyle='-',
    linewidth=2,
    color='green'
)

# Menampilkan nilai di setiap titik
for x, y in zip(timeline.index, timeline.values):
    plt.text(
        x,
        y + 5,
        str(y),
        ha='center',
        fontsize=8
    )

plt.title("Customer Activity Timeline")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Orders")

plt.xticks(range(24))
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

# Insight
Aktivitas pembelian seiring bertambahnya order pelanggan

# 5. Jam Paling Ramai — Bar Chart
Tujuan: Mengetahui jam dengan traffic order tertinggi agar perusahaan dapat mengoptimalkan server, pengiriman, dan promosi.

```python
import seaborn as sns
import matplotlib.pyplot as plt

jam_ramai = (
    df.groupby('order_hour_of_day')['order_id']
      .nunique()
      .reset_index(name='total_orders')
)

peak_hour = jam_ramai['total_orders'].idxmax()

colors = ['tab:blue'] * len(jam_ramai)
colors[peak_hour] = 'tab:red'

plt.figure(figsize=(12,6))

ax = sns.barplot(
    data=jam_ramai,
    x='order_hour_of_day',
    y='total_orders',
    palette=colors
)

for container in ax.containers:
    ax.bar_label(container, fmt='%d', padding=3)

plt.title('Jam Paling Ramai Berdasarkan Jumlah Order', fontsize=14, fontweight='bold')
plt.xlabel('Jam')
plt.ylabel('Jumlah Order')

plt.tight_layout()
plt.show()
```

# Insight
* Puncak transaksi pukul 12.00.

Rekomendasi
* Tingkatkan kapasitas sistem dan stok pada jam tersebut.




# 6. 10 Produk Terlaris — Bar Chart Vertikal
Tujuan: Mengetahui produk paling sering dibeli untuk strategi stok, bundling, dan promosi.

```python
# Top 10 produk berdasarkan jumlah pembelian
top_produk = (
    df.groupby('product_name')['order_id']
      .count()
      .sort_values(ascending=False)
      .head(10)
      .reset_index(name='total_pembelian')
)

# Style
sns.set_style("whitegrid")

plt.figure(figsize=(12, 6))

ax = sns.barplot(
    data=top_produk,
    x='product_name',
    y='total_pembelian'
)

# Label nilai di atas bar
for container in ax.containers:
    ax.bar_label(container, fmt='%d', padding=3, fontsize=9)

plt.title('10 Produk Terlaris', fontsize=14, fontweight='bold')
plt.xlabel('Nama Produk')
plt.ylabel('Jumlah Pembelian')

plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

# Insight
* Banana menjadi produk paling laris.

Rekomendasi
* Prioritaskan stok produk tersebut.

# 7. Reorder vs Non Reorder — Bar cart
Tujuan: Melihat perbandingan jumlah produk yang di-reorder dan tidak untuk memahami customer retention.

```python
df['reorder_status'] = df['reordered'].map({
    0: 'Non Reorder',
    1: 'Reorder'
})

plt.figure(figsize=(8,5))

ax = sns.countplot(
    data=df,
    x='reorder_status'
)

for container in ax.containers:
    ax.bar_label(container)

plt.title('Reorder vs Non Reorder')
plt.xlabel('Status Pembelian')
plt.ylabel('Jumlah Produk')

plt.show()
```

# Insight
* Sebagian besar produk dibeli ulang.

Rekomendasi
* Buat loyalty program.

# 8. Hubungan Barang Dimasukkan dan Barang Re-order
Tujuan: Menganalisis apakah urutan produk dimasukkan ke keranjang mempengaruhi kemungkinan produk di-reorder.

```python
import matplotlib.pyplot as plt
import seaborn as sns

cart_reorder = (
    df.groupby('add_to_cart_order')['reordered']
      .mean()
      .reset_index()
)

plt.figure(figsize=(12, 6))

sns.scatterplot(
    data=cart_reorder,
    x='add_to_cart_order',
    y='reordered'
)

plt.title('Peluang Reorder Berdasarkan Urutan Produk dalam Keranjang',
          fontsize=14, fontweight='bold')
plt.xlabel('Urutan Produk Ditambahkan ke Keranjang')
plt.ylabel('Rata-rata Reorder')

plt.tight_layout()
plt.show()
```

# Insight
* Produk yang masuk lebih awal cenderung sering direorder.

Rekomendasi
* Gunakan sebagai dasar sistem rekomendasi.
